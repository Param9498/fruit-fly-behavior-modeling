```python
import numpy as np
import os
```

```python
data_folder = '../data'
```

```python
with open(os.path.join(data_folder, 'all_data.npy'), 'rb') as f:
    np_data = np.load(f)
```

```python
np_data.shape
```

```python
import matplotlib.pyplot as plt

import statsmodels.api as sm

from statsmodels.tsa.stattools import acf, ccf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
lag = 20
index = 200
plt.plot(np_data[index, :, 0])
plot_acf(x=np_data[index, :, 0], lags=lag, title="ACF w_t")
plt.show()

#TODO: plot the pacf of the data
plt.figure()
sm.graphics.tsa.plot_pacf(np_data[index, :, 0], lags=40)
plt.show()
```
