```python
import pandas as pd
import numpy as np

import os
```

```python
data_folder = '../data'
```

```python
x = pd.read_csv(os.path.join(data_folder, 'x.csv'), header=None)
y = pd.read_csv(os.path.join(data_folder, 'y.csv'), header=None)
fvel = pd.read_csv(os.path.join(data_folder, 'fvel.csv'), header=None)
avel = pd.read_csv(os.path.join(data_folder, 'avel.csv'), header=None)
theta = pd.read_csv(os.path.join(data_folder, 'theta.csv'), header=None)
utheta = pd.read_csv(os.path.join(data_folder, 'utheta.csv'), header=None)
whichfly = pd.read_csv(os.path.join(data_folder, 'whichfly.csv'), header=None)
upwindvel = pd.read_csv(os.path.join(data_folder, 'upwindvel.csv'), header=None)
```

```python
print(f''' SHAPES:
x \t\t :- \t{x.shape}
y \t\t :- \t{y.shape} 
fvel \t\t :- \t{fvel.shape} 
avel \t\t :- \t{avel.shape} 
theta \t\t :- \t{theta.shape} 
utheta \t\t :- \t{utheta.shape} 
whichfly \t :- \t{whichfly.shape} 
upwindvel \t :- \t{upwindvel.shape} 
''')
```

```python
x = x.iloc[:3500]
y = y.iloc[:3500]
fvel = fvel.iloc[:3500]
avel = avel.iloc[:3500]
theta = theta.iloc[:3500]
utheta = utheta.iloc[:3500]
whichfly = whichfly.iloc[:3500]
upwindvel = upwindvel.iloc[:3500]
```

```python
assert x.shape == y.shape == fvel.shape == avel.shape == theta.shape == utheta.shape == whichfly.shape == upwindvel.shape
```

```python
shape = x.shape
```

### JSON formatting

```python
data = []
for trial in range(shape[1]):
    data.append({
        'trial': int(trial),
        'x': x[trial].values.tolist(),
        'y': y[trial].values.tolist(),
        'fvel': fvel[trial].values.tolist(),
        'avel': avel[trial].values.tolist(),
        'theta': theta[trial].values.tolist(),
        'utheta': utheta[trial].values.tolist(),
        'whichfly': whichfly[trial].values.tolist()[0],
        'upwindvel': upwindvel[trial].values.tolist()
    })
```

```python
import json
with open(os.path.join(data_folder, 'all_data.json'), 'w') as f:
    json.dump(data, f)
```

```python
unique_flies = []
for col in range(shape[1]):
    unique_flies.append(whichfly[col].unique()[0])

unique_flies = list(set(unique_flies))
# unique_flies  
```

```python
for fly in unique_flies:
    fly_data = []
    for d in data:
        if d['whichfly'] == fly:
            fly_data.append(d)

    with open(os.path.join(data_folder, 'per_fly', f'fly_{fly}.json'), 'w') as f:
        json.dump(fly_data, f)
```

### NUMPY FORMATTING FROM JSON

```python
with open(os.path.join(data_folder, 'all_data.json'), 'r') as f:
    data = json.load(f)
```

```python
data[0]
```

```python
np_data = []
for trial_data in data:
    np_data.append([
#         [trial_data['trial']]*len(trial_data['x']),
        trial_data['x'],
        trial_data['y'],
        trial_data['fvel'],
        trial_data['avel'],
        trial_data['theta'],
        trial_data['utheta'],
        [trial_data['whichfly']] * len(trial_data['x']),
        trial_data['upwindvel']
    ])
```

```python
np_data = np.array(np_data).transpose(0, 2, 1)
```
