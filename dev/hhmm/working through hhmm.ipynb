```python
catformodel = categories[:,0]
fformodel = fvel[:,0]
aformodel = avel[:,0]

for ii in range(1,327):
    catformodel = np.concatenate((catformodel,categories[:498,ii]))
    fformodel = np.concatenate((fformodel,fvel[:498,ii]))
    aformodel = np.concatenate((aformodel,avel[:498,ii]))


print(len(catformodel))
print(len(aformodel))
print(len(fformodel))

formodel = np.reshape(catformodel, (-1,1))
```

```python
sampling_freq = 50
from hmmlearn import hmm
from hmmlearn.base import ConvergenceMonitor

np.random.seed(42)

highlevelmodel = hmm.GaussianHMM(n_components=6,  verbose=True, n_iter=10000)

highlevelmodel.monitor_ = ConvergenceMonitor(tol=0.01, n_iter=10000, verbose=True)
print(highlevelmodel.monitor_)
```

```python
#set up the data for the lower level states 

a = len(aformodel)

for kk in range(1, 7):
    valsa =[]
    valsf = []
    lengths = []
    count = 0 
    for ii in range(a):
        if catformodel[ii] == kk:
            valsa.append(aformodel[ii])
            valsf.append(fformodel[ii])
            count += 1
        else:
            if count > 0:
                lengths.append(count)
            count = 0

    valsa = np.array([valsa])
    valsf = np.array([valsf])
    
    vals_af = np.concatenate([valsa, valsf])
    
#     print(valsa.shape, valsf.shape)
#     print(sum(lengths))
    if kk == 1:
        state1model = hmm.GaussianHMM(n_components=10,  verbose=True, n_iter=10000)
        state1model.monitor_ = ConvergenceMonitor(tol=0.01, n_iter=10000, verbose=True)
        state1model.fit(vals_af.T, lengths)
        
    elif kk == 2:
        state2model = hmm.GaussianHMM(n_components=10,  verbose=True, n_iter=10000)
        state2model.monitor_ = ConvergenceMonitor(tol=0.01, n_iter=10000, verbose=True)
        state2model.fit(vals_af.T, lengths)
    elif kk == 3:
        state3model = hmm.GaussianHMM(n_components=10,  verbose=True, n_iter=10000)
        state3model.monitor_ = ConvergenceMonitor(tol=0.01, n_iter=10000, verbose=True)
        state3model.fit(vals_af.T, lengths)
    elif kk == 4:
        state4model = hmm.GaussianHMM(n_components=10,  verbose=True, n_iter=10000)
        state4model.monitor_ = ConvergenceMonitor(tol=0.01, n_iter=10000, verbose=True)
        state4model.fit(vals_af.T, lengths)
    elif kk == 5:
        state5model = hmm.GaussianHMM(n_components=10,  verbose=True, n_iter=10000)
        state5model.monitor_ = ConvergenceMonitor(tol=0.01, n_iter=10000, verbose=True)
        state5model.fit(vals_af.T, lengths)
    elif kk == 6:
        state6model = hmm.GaussianHMM(n_components=10,  verbose=True, n_iter=10000)
        state6model.monitor_ = ConvergenceMonitor(tol=0.01, n_iter=10000, verbose=True)
        state6model.fit(vals_af.T, lengths)
```

```python
state_models = {
    1: state1model,
    2: state2model,
    3: state3model,
    4: state4model,
    5: state5model,
    6: state6model,
}
```

```python
series = highlevelmodel.sample(499, random_state=1234)[0]
sampled_categories =  series[:, 0]
```

```python
sampled_categories = np.round(sampled_categories)
```

```python
sampled_categories
```

```python
previous_cat = sampled_categories[0]
grouped_categories = []
intermediate_categories = []
for element in sampled_categories:
    if element == previous_cat:
        intermediate_categories.append(int(element))
    else:
#         intermediate_categories.append(element)
        grouped_categories.append(intermediate_categories)
        intermediate_categories = [int(element)]
    previous_cat = element
        
```

```python
grouped_categories
```

```python
observations = []
i = 0
for group in grouped_categories:
    
    cat = group[0]
    
    if cat in [5, 6]:
        continue
    
    if i == 0:
        obs = state_models[cat].sample(len(group))[0]
        for o in obs:
            observations.append(o)
    else:
        last_observation = observations[-1]
        
        init_prob = state_models[cat].startprob_
        
#         print(state_models[cat].predict_proba([last_observation]).shape)
        
        state_models[cat].startprob_ = state_models[cat].predict_proba([last_observation])[0]
        
        obs = state_models[cat].sample(len(group))[0]
        
        for o in obs:
            observations.append(o)
        
        state_models[cat].startprob_ = init_prob
        
    i += 1
```
