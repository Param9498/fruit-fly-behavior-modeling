```python
# Method 1

num_samples = 30
num_timesteps = 499
feature_dimensions = 2

all_observations = []

for j in range(num_samples):
    
    # Highter Level HMM sampling
    series = highlevelmodel.sample(num_timesteps)[0]
    sampled_categories =  series[:, 0]
    sampled_categories = np.round(sampled_categories)
    
    # Higher level HMM samples grouping
    previous_cat = sampled_categories[0]
    grouped_categories = []
    intermediate_categories = []

    ind = 0
    for element in sampled_categories:
        if element == previous_cat:
            intermediate_categories.append(int(element))
        else:
            grouped_categories.append(intermediate_categories)
            intermediate_categories = [int(element)]
            
        if ind == len(sampled_categories) - 1:
            grouped_categories.append(intermediate_categories)
        previous_cat = element

        ind += 1
    
    
    # Lower Level HMM sampling
    observations = []
    i = 0
    for group in grouped_categories:

        cat = group[0]

        if i == 0:
            obs = state_models[cat].sample(len(group))[0]
            for o in obs:
                observations.append(list(o))
        else:
            last_observation = observations[-1]
            init_prob = state_models[cat].startprob_
            state_models[cat].startprob_ = state_models[cat].predict_proba([last_observation])[0]
            obs = state_models[cat].sample(len(group))[0]
            for o in obs:
                observations.append(list(o))
            state_models[cat].startprob_ = init_prob

        i += 1
    all_observations.append(np.array(observations))
```

```python
from pathlib import Path

sample_save_path = "samples/method_1/"
Path(sample_save_path).mkdir(parents=True, exist_ok=True)

for i in range(num_samples):
    sample_df = pd.DataFrame(np.array(all_observations[i]), columns=['avel', 'fvel'])
    sample_df['avel'] = sample_df['avel']/50
    print(sample_df)
    sample_df.to_csv(os.path.join(sample_save_path, f"{i+1}.csv"), index=False)
```
